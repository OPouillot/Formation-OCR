```python
import pandas as pd
import requests
import json
```

```python
url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {"ingr":"['champagne']"}

headers = {
	"X-RapidAPI-Key": "0ef763ce33msha9485b34fbf1b2ap156630jsnc364b2604352",
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
data = response.json()

food_data = []

for hint in data['hints']:
    food_data.append(hint['food'])

df = pd.DataFrame(food_data)
df.head()
```

```python
filtered_df = df.loc[:10, ['foodId', 'label', 'category', 'foodContentsLabel', 'image']]
filtered_df
```
